# Data Wrangling

```python
customers.shape
```

```python
customers.head()
```

```python
# rename First Name column
customers.rename(columns = {'First Name' : 'first_name'}, inplace = True)
```

```python
# rename Surnam column
customers.rename(columns = {'Surnam' : 'surname'}, inplace = True)
```

```python
# rename Gender column
customers.rename(columns = {'Gender' : 'gender'}, inplace = True)
```

```python
# rename Age column
customers.rename(columns = {'Age': 'age'}, inplace = True)
```

```python
# rename n_dependents column
customers.rename(columns = {'n_dependents' : 'number_of_dependents'}, inplace = True)
```

```python
# rename fam_status column
customers.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```

```python
#rename STATE column
customers.rename(columns = {'STATE' : 'state'}, inplace = True)
```

```python
customers.head()
```

```python
# find columns mixed-type columns
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# find columns with missing values
customers.isnull().sum()
```

```python
# change first_name data type to string
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# find duplicate values
customers_dups = customers[customers.duplicated()]
```
