## 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform( ) and loc ( ) functions.

```python
# create max_order coloumn
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# create 'loyal customer' loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] >40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# create 'regular customer' loyalty flag
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# create 'new customer' loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category. 

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

## 6. Create a spending flag for each user based on the average price across all their orders.

```python
# create mean_price column
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# create 'low spender' spending flag
ords_prods_merge.loc[ords_prods_merge['mean_price'] <10, 'spending_flag'] = 'Low spender'
```

```python
# create 'high spender' spending flag
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## 7. Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the "days_since_prior_order" column.

```python
# create median_days_since_last_order column
ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# create ''non-frequent customer' frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# create 'regular customer' frequency flag
ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] <= 20) & (ords_prods_merge['median_days_since_last_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# create 'frequent custoemr' frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```
